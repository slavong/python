```python
import matplotlib.pyplot as plt

year = [1,2,3,4,5]
pop = [3.4,5.3,6.2,6.9,7.1]

plt.plot(year, pop)

plt.xlabel('Year')
plt.ylabel('Population')

plt.title('World Population by Year')

plt.yticks([0,2,4,6,8,10], ['0', '2B', '4B', '6B', '8B', '10B'])

plt.grid()

plt.show()
```
