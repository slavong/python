# Aggregating Data Frames

```python
df["avg_temp_c"].mean()
```

```python
df["avg_temp_c"].median()
df["avg_temp_c"].mode()
df["avg_temp_c"].min()
df["avg_temp_c"].max()
df["avg_temp_c"].var()
df["avg_temp_c"].std()
df["avg_temp_c"].sum()
df["avg_temp_c"].quantile()

print(df["avg_temp_c"].quantile())
print(df["avg_temp_c"].quantile(.3))
```

```python
def pct30(column):
    return column.quantile(0.3)

df["avg_temp_c"].agg(pct30)
```

```python
df[["avg_temp_c", "avg_temp_f"]].agg(pct30)
```

```python
def pct40(column):
    return column.quantile(0.4)

df[["avg_temp_c", "avg_temp_f"]].agg([pct30, pct40])
```

```python
df["avg_temp_c"].cumsum()
```

```python
df["avg_temp_c"].cummin()
df["avg_temp_c"].cummax()
#df["avg_temp_c"].cumprod()  # leads to error
df[df["city"] == "Berlin"]["avg_temp_c"].cumprod()
```

### Pivot Tables

# Slicing and Indexing DataFrames

```python
array = [
    ("Bella", "labrador", "brown", 56, 25),
    ("Charlie","poodle","black", 43, 23),
    ("Lucy","chow chow","brown", 46, 22),
    ("Cooper", "schnauzer", "grey", 49, 17),
    ("Max", "labrador", "black", 59, 29),
    ("Stella", "chihuahua", "tan", 18, 2),
    ("Bernie", "st.bernard", "white", 77, 74),
    ]
dogs = pd.DataFrame(array, columns=["name", "breed", "color", "height_cm", "weight_kg"])
dogs.set_index("name")
```

```python
dogs.reset_index(drop=True)
```

```python
dogs.reset_index(drop=False)
```

```python
dogs[dogs["name"].isin(["Bella","Stella"])]
# or
dogs.set_index("name").loc[["Bella", "Stella"]]
```

```python
dogs.set_index("breed").loc["labrador"]
```

```python
dogs.set_index(["breed","color"])
```

```python
dogs.set_index(["breed","color"]).loc[["labrador", "chihuahua"]]
```

```python
dogs.set_index(["breed","color"]).loc[[("labrador", "black"), ("chihuahua", "tan")]]
```

```python
dogs.set_index(["breed","color"]).sort_index()
```

```python
dogs.set_index(["breed","color"]).sort_index(level=["color", "breed"], ascending=[False, True])
```
