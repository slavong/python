```python
import numpy as np
EXPERIMENTS = 100
THROWS = 10
np.random.seed(42)
final_tails = []
for x in range(EXPERIMENTS):
    tails = [0]
    for x in range(THROWS):
        coin = np.random.randint(0,2)
        tails.append(tails[x] + coin)
    final_tails.append(tails[-1])
print(final_tails)
```

```python
import matplotlib.pyplot as plt

plt.hist(final_tails, bins=10)

plt.show()
```

```python
stats = [0 for i in range(THROWS+1)]
for result in final_tails:
    stats[result] += 1
stats
```

```python
import matplotlib.pyplot as plt

plt.bar(range(THROWS+1), height=stats)

plt.show()
```
